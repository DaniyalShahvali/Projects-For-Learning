```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Extracting features and target as our x and y data
x_data = diabetes_data.drop('Outcome', axis = 1)
y_data = diabetes_data['Outcome']
```

```python
# splitting data to train and test subgroup
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, random_state= 7) 
```

```python
print(f'x_train shape: {x_train.shape}')
print(f'y_train shape: {y_train.shape}')
```

```python
print(f'x_test shape: {x_test.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors= 1, metric='euclidean')
```

```python
knn.fit(x_train, y_train)
```

```python
knn.score(x_test, y_test)
```

```python
# Finding the optimal K value (Elbow method)
K_range = range(1, 31, 2)
scores = []

for k in K_range:
    KNN_model = KNeighborsClassifier(n_neighbors= k, metric= 'euclidean')
    KNN_model.fit(x_train, y_train)
    score = KNN_model.score(x_test, y_test)
    scores.append(score)
```

```python
# Plot the result
plt.figure(figsize= (5,5))
plt.plot(K_range, scores, 'bo-', alpha = 0.7, linewidth = 2, markersize = 6)
plt.xlabel('Value of K for KNN')
plt.ylabel('Scores based on value of K')
plt.title('Optimal value for K')
plt.grid(True, alpha = 0.3)

optimal_k = K_range[np.argmax(scores)]
best_score = max(scores)
print(f"Optimal k: {optimal_k}")
print(f"Best score: {round(best_score, 3)}")

plt.axvline(x= optimal_k, color = 'red', linestyle = '--', alpha = 0.7, label = f'Optimal K = {optimal_k}')
plt.legend()
plt.show()
```
